```python
import pandas as pd
import matplotlib.pyplot as plt
```

```python
file = '/content/drive/MyDrive/DS Intern/Task 01/world population.xls'

df = pd.read_excel(file)
```

```python
df.head(100)
```

```python
df.describe()
```

```python
df.info()
```

```python
india_data = df[df['Country Name'] == 'India']

years = list(map(str, range(1980, 2022)))
india_population = india_data[years].T
india_population.columns = ['Population']

plt.figure(figsize=(12, 6))
plt.bar(india_population.index, india_population['Population'], color='skyblue')
plt.xlabel('Year')
plt.ylabel('Population')
plt.title('Population Growth of India (1980 - 2021)')
plt.xticks(rotation=45, ha='right')
plt.grid(axis='y')
plt.show()
```
